```python
import pandas as pd
import requests
!pip install bs4
from bs4 import BeautifulSoup
```

```python
#----To store phone and price created a list
model=[]
price=[]
Configuration=[]


pgnumber=int(input("Enter Page Number : "))
for i in range(1,pgnumber+1):
    url="https://www.flipkart.com/search?q=iphone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_3_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_2_3_na_na_na&as-pos=2&as-type=HISTORY&suggestionId=iphone&requestId=b4e64f4f-1fc6-4a52-9246-06df6b7fffba&page="+str(i)
    #-----Number of pages changes according
    re=requests.get(url)#-------------requests for this url
    content=BeautifulSoup(re.content,'html.parser')#----getting content in the form of html
    #print(content)
    ph_name=content.find_all('div',{'class':'_4rR01T'})
    ph_price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
    print("Total Iphones in page : "+str(i))
    print(len(ph_name))
    ph_info=content.find_all('div',{'class':'fMghEO'})    
    
    
    
#---To append the ph_name data
    for i in ph_name:
        model.append(i.text)#-----all pg ph_name append here
        
#---To append price data
    for i in ph_price:
        price.append(i.text)#----to get just text
        
#---To append info
    for i in ph_info:
        Configuration.append(i.text)
```

```python
for i in model:
    print("Model  :   ",i)
```

```python
data={"Iphone Model":model,"Price":price,"Configuration":Configuration}
df=pd.DataFrame.from_dict(data, orient='index')
df=df.transpose()
print(df)
```

```python
dff=pd.DataFrame(data,columns=['Iphone Model','Price','Configuration'])

dff
```
